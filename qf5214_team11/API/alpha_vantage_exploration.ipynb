```python
from typing import *
```

```python
def get_data(function:str, 
             symbols:List[str], 
             apikey:str,
             interval:str,
             month: str):
    data_list = []
    for symbol in symbols:
        url = 'https://www.alphavantage.co/query?function={}&symbol={}&month={}&apikey={}&interval={}'.format(function, symbol, month, apikey, interval)
        r = requests.get(url)
        data = r.json()
        data_list.append(data)
    return data_list   
```

```python
stock_list = get_data('TIME_SERIES_INTRADAY', 
         symbols=['IBM','TSLA'], 
         apikey='TFV8EPDGH6TGK363',
         interval='1min',
         month='2024-03')
```

```python
for item in stock_list:
    print(item)
```
